```r
#B)
regression = glm(Direction ~ Lag1 + L
                 ag2 + Lag3 + Lag4 + Lag5 + Volume, family = "binomial", data = Weekly)
summary(regression)
#only value showing statistical relevance is lag 2 due to its o value (less than .05)
```

```r
#C)
#Correct prediction on the training data is 56.1% 

p <- predict(regression, Weekly, type="response")
predicted <- rep("Down", 1089)
predicted[p > 0.5] <- "Up"
table(predicted, Weekly$Direction)
```

```r
#D 
#fit the model
train <- (Weekly$Year < 2009)
t_w <- Weekly[!train, ]
t_dir <- Weekly$Direction[!train]
regression_2 <- glm(Direction ~ Lag2, data = Weekly, family = binomial, subset = train)
summary(regression_2)
```

```r
#confusion matrix
p_2 <- predict(regression_2, t_w, type="response")
pred_2 <- rep("Down", 104)
pred_2[p_2 > 0.5] <- "Up"
table(pred_2, t_w$Direction)
```
